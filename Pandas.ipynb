1. The name Pandas is derived from the word Panel Data – an Econometrics from Multidimensional data.
2. Used for data manipulation and analysis (load, prepare, manipulate, model, and analyze).
3. Offers data structures and operations for manipulating numerical tables and time series.
4. Written in  Python, Cython, C. 
5. Developed by Wes McKinney in 2008.
6. Pandas deal with three data structures: Series-1D, Dataframe-2D, Panel-3D (all these data structures are built on top of numpy array, that's why they are fast.


```python
# Series
# It is one-dimensional labeled array capable of holding data of any type.
# Has four parameters - pandas.Series( data, index, dtype, copy).
# data(data takes various forms like ndarray, list, constants), index(must be unique and hashable, same length as data. Default np.arrange(n) if no index is passed), dtype(dtype is for data type. If None, data type will be inferred), copy(Copy data. Default False)
import pandas as pd
s = pd.Series()                                         #Empty series

import numpy as np
data = np.array(['a','b','c','d'])                      #Creating a Series from ndarray
s = pd.Series(data)
print ("s = ",s)

data = {'a' : 0., 'b' : 1., 'c' : 2.}                   #Creating series from dictionary
s1 = pd.Series(data)
print ("s1 =",s1)

data = {'a' : 0., 'b' : 1., 'c' : 2.}                   #Creating series from dictionary
s2 = pd.Series(data,index=['b','c','d','a'])
print ("s2 = ",s2)

s3 = pd.Series(5, index=[0, 1, 2, 3])                   #Creating series from scalar
print ("s3 = ", s3)

print (s2[0])                                             #Accessing series (1st element)
print (s2[:3])                                            #Retrieving first three elements
print (s2[-3:])                                           #Retrieving last three elements 
print (s2['a'])                                           #Retrieving elements using labels 
print (s2[['a','c','d']])                                  #Retrieving multiple elements.
```

```python
# DATAFRAME
# Data is aligned in a tabular fashion in rows and columns.
# Has 5 parameters - pandas.DataFrame( data, index, columns, dtype, copy)
# data(data takes various forms like ndarray, series, map, lists, dict, constants and also another DataFrame), index(For the row labels, the Index to be used for the resulting frame is Optional Default np.arange(n) if no index is passed), columns(For column labels, the optional default syntax is - np.arange(n). This is only true if no index is passed), dtype(Data type of each column), copy(used for copying of data, if the default is False)
import pandas as pd
df = pd.DataFrame()                                   #Empty dataframe
print (df , "-----")

data = [1,2,3,4,5]                                    #Create dataframe from list
df1 = pd.DataFrame(data)
print (df1 , "------") 

data = [['Alex',10],['Bob',12],['Clarke',13]]
df2 = pd.DataFrame(data,columns=['Name','Age'])
print (df2, "------")

data = [['Alex',10],['Bob',12],['Clarke',13]]
df3 = pd.DataFrame(data,columns=['Name','Age'],dtype=float)
print (df3, "-------")

data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}            # Creating DataFrame from Dict of ndarrays / Lists
df4 = pd.DataFrame(data)
print (df4, "------")

data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df5 = pd.DataFrame(data, index=['rank1','rank2','rank3','rank4'])                # index parameter assigns an index to each row.
print (df5, "------")

data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]    
df6 = pd.DataFrame(data)    
df7 = pd.DataFrame(data, index=['first', 'second'], columns=['a', 'b'])          # With two column indices, values same as dictionary keys
print (df6, "-------")
print (df7, "------")

d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']),                        # DataFrame from Dict of Series
   'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
df8 = pd.DataFrame(d)
print (df8, "------")

print (df8 ['one'], "------")                                                                 # Column selection

# Adding a new column to an existing DataFrame object with column label by passing new series

print ("Adding a new column by passing as Series:")
df8['three']=pd.Series([10,20,30],index=['a','b','c'])
print (df8, "------")

print ("Adding a new column using the existing columns in DataFrame:")
df8['four']=df8['one']+df8['three']

print (df8, "------")

# Deletion of column
# using del function
print ("Deleting the first column using DEL function:")
del df8['one']
print (df8, "------")

# using pop function
print ("Deleting another column using POP function:")
df8.pop('two')
print (df8,"------")

print (df8.loc['b'], "--------")                                                               # Row selection by passing row label
print (df8.iloc[2], "--------")                                                                # row selection by passing index to an iloc function
print (df8[2:4], "--------")                                                                   # row slicing

df = pd.DataFrame([[1, 2], [3, 4]], columns = ['a','b'])
df2 = pd.DataFrame([[5, 6], [7, 8]], columns = ['a','b'])

df = df.append(df2)                                                                    # Addition of rows
print (df,"-----")

df = df.drop(0)                                                                       # Deletion of row
print (df, "-------")
```

##FUNCTIONS ##
1. Table wise Function Application: pipe()
2. Row or Column Wise Function Application: apply()
3. Element wise Function Application: applymap()



```python
# Table-wise Function Application
def adder(ele1,ele2):
   return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print (df.pipe(adder,2))
print (" ")
print (df.apply(np.mean))


# Row or Column Wise Function Application
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.apply(np.mean)
print (" ")
print (df.apply(np.mean))

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.apply(lambda x: x.max() - x.min())
print(" ")
print (df.apply(np.mean))

# Element Wise Function Application
# Not all functions can be vectorized (neither the NumPy arrays which return another array nor any value), the methods applymap() on DataFrame and analogously map() on Series accept any Python function taking a single value and returning a single value.
df['col1'].map(lambda x:x*100)
print (" ")
print (df.apply(np.mean))

df.applymap(lambda x:x*100)
print(" ")
print (df.apply(np.mean))
```

```python
# Reindexing : changes the row labels and column labels of a DataFrame

N=20

df = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),
   'x': np.linspace(0,stop=N-1,num=N),
   'y': np.random.rand(N),
   'C': np.random.choice(['Low','Medium','High'],N).tolist(),
   'D': np.random.normal(100, 10, size=(N)).tolist()
})

#reindex the DataFrame
df_reindexed = df.reindex(index=[0,2,5], columns=['A', 'C', 'B'])

print (df_reindexed)


#Reindex to Align with Other Objects
df1 = pd.DataFrame(np.random.randn(10,3),columns=['col1','col2','col3'])
df2 = pd.DataFrame(np.random.randn(7,3),columns=['col1','col2','col3'])

df1 = df1.reindex_like(df2)                       #take an object and reindex its axes to be labeled the same as another object. 
print("--------")
print (df1)

#Filling while ReIndexing
# reindex() takes an optional parameter "method" which is a filling method with values as follows −pad/ffill (Fill values forward), bfill/backfill (Fill values backward), nearest (Fill from the nearest index values)

# Padding NAN's
print (df2.reindex_like(df1))

# Now Fill the NAN's with preceding Values
print ("Data Frame with Forward Fill:")
print (df2.reindex_like(df1,method='ffill'))

# Limits on Filling while Reindexing
# The limit argument provides additional control over filling while reindexing. Limit specifies the maximum count of consecutive matches. 

print("----------")
print (df2.reindex_like(df1,method='ffill',limit=1))




# Renaming :The rename() method allows you to relabel an axis based on some mapping (a dict or Series) or an arbitrary function.

df1 = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
print (df1)

print ("After renaming the rows and columns:")
print (df1.rename(columns={'col1' : 'c1', 'col2' : 'c2'},
index = {0 : 'apple', 1 : 'banana', 2 : 'durian'}))
```

```python
# ITERATION
# To iterate over the rows of the DataFrame, we can use the following functions −
# 1. iteritems() − to iterate over the (key,value) pairs
# 2. iterrows() − iterate over the rows as (index,series) pairs
# 3. itertuples() − iterate over the rows as namedtuples

#ITERITEMS() : Iterates over each column as key, value pair with label as key and column value as a Series object.
df = pd.DataFrame(np.random.randn(4,3),columns=['col1','col2','col3'])
for key,value in df.iteritems():
    print (key,value)
print("--------")
    
#ITERROWS() : iterrows() returns the iterator yielding each index value along with a series containing the data in each row.
for row_index,row in df.iterrows():
    print(row_index,row)
print("--------")

#itertuples() : itertuples() method will return an iterator yielding a named tuple for each row in the DataFrame. The first element of the tuple will be the row’s corresponding index value, while the remaining values are the row values.
for row in df.itertuples():
    print (row)
```

```python
# SORTING
# Two kinds : 1. by label , 2. by actual value

unsorted_df=pd.DataFrame(np.random.randn(10,2),index=[1,4,6,2,3,5,9,8,0,7],columns=['col2','col1'])
print (unsorted_df)
print("-------")

# BY LABEL
# by passing the axis arguments and the order of sorting in sort_index(). By default, sorting is done on row labels in ascending order.
sorted_df=unsorted_df.sort_index()
print (sorted_df)
print("-------")
sorted_df1 = unsorted_df.sort_index(ascending=False)                     #Order of sorting
print (sorted_df1)
print("--------")
sorted_df2=unsorted_df.sort_index(axis=1)                                # Sort columns
print (sorted_df2)
print("--------")

# BY VALUE
# Like index sorting, sort_values() is the method for sorting by values. It accepts a 'by' argument which will use the column name of the DataFrame with which the values are to be sorted.
unsorted_df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
sorted_df = unsorted_df.sort_values(by='col1')
print (sorted_df)
print("--------")
sorted_df1 = unsorted_df.sort_values(by=['col1','col2'])
print (sorted_df1)
print ("--------")

#Sorting Algorithm
sorted_df2 = unsorted_df.sort_values(by='col1' ,kind='mergesort')
print (sorted_df2)
```

```python
# STATISTICAL FUNCTIONS 

# Percent_change : compares every element with its prior element and computes the change percentage.
# By default, the pct_change() operates on columns; if you want to apply the same row wise, then use axis=1() argument.
s = pd.Series([1,2,3,4,5,4])
print (s.pct_change())
df = pd.DataFrame(np.random.randn(5, 2))
print (df.pct_change())
print("-------")

# Covariance :Covariance is applied on series data. The Series object has a method cov to compute covariance between series objects. NA will be excluded automatically.
s1 = pd.Series(np.random.randn(10))
s2 = pd.Series(np.random.randn(10))
print (s1.cov(s2))
print("-------")

frame = pd.DataFrame(np.random.randn(10, 5), columns=['a', 'b', 'c', 'd', 'e'])
print (frame['a'].cov(frame['b']))                         # Covariance method when applied on a DataFrame, computes cov between all the columns.
print (frame.cov())
print("--------")

# Correlation : Shows the linear relationship between any two array of values (series). There are multiple methods to compute the correlation like pearson(default), spearman and kendall.
print (frame.corr())                            # If any non-numeric column is present in the DataFrame, it is excluded automatically.
print("--------")

# Data Ranking : Ranking for each element in the array of elements. In case of ties, assigns the mean rank.
# Rank optionally takes a parameter ascending which by default is true; when false, data is reverse-ranked, with larger values assigned a smaller rank.
# Rank supports different tie-breaking methods, specified with the method parameter −
# 1. average − average rank of tied group
# 2. min − lowest rank in the group
# 3. max − highest rank in the group
# 4. first − ranks assigned in the order they appear in the array

s = pd.Series(np.random.np.random.randn(5), index=list('abcde'))
s['d'] = s['b'] # so there's a tie
print (s.rank())
print("---------")

```

```python
# WINDOW FUNCTIONS (.rolling , .expanding, .ewm)
# Window functions are majorly used in finding the trends within the data graphically by smoothing the curve. If there is lot of
# variation in the everyday data and a lot of data points are available, then taking the samples and plotting is one method and
# applying the window computations and plotting the graph on the results is another method. By these methods, we can smooth the 
# curve or the trend.


# .rolling() Function : Can be applied on a series of data. Specify the window=n argument and apply the appropriate statistical function on top of it.
df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2000', periods=10),
   columns = ['A', 'B', 'C', 'D'])
print (df.rolling(window=3).mean())
print("---------")

# .expanding() Function : Can be applied on a series of data. Specify the min_periods=n argument and apply the appropriate statistical function on top of it.
print (df.expanding(min_periods=3).mean())
print("---------")

# .ewm() Function : Can be applied on a series of data. Specify any of the com, span, halflife argument and apply the appropriate statistical function on top of it. It assigns the weights exponentially.
print (df.ewm(com=0.5).mean())

```

```python
#AGGREGATIONS

df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2000', periods=10),
   columns = ['A', 'B', 'C', 'D'])
print (df)
print("--------")
r = df.rolling(window=3,min_periods=1)
print (r.aggregate(np.sum))                                  # Apply aggregation on whole data frame
print("---------")
print (r['A'].aggregate(np.sum))                             # Apply aggregation on single column 
print("---------")
print (r[['A','B']].aggregate(np.sum))                       # Apply aggregation on multiple columns
print("----------")
print (r['A'].aggregate([np.sum,np.mean]))                   # Apply multiple function of single column of dataframe
print("---------")
print (r[['A','B']].aggregate([np.sum,np.mean]))             # Apply multiple functions on multiple columns
print("----------")
print (r.aggregate({'A' : np.sum,'B' : np.mean}))            # Apply Different Functions to Different Columns of a Dataframe
```

```python
# HANDLING NULL VALUES 

df = pd.DataFrame(np.random.randn(5, 3), index=['a', 'c', 'e', 'f',
'h'],columns=['one', 'two', 'three'])

df = df.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
print (df)
print("---------")
print (df['one'].isnull())
print("---------")
print (df['one'].notnull())
print("---------")


# CALCULATIONS WITH MISSING DATA
# When summing data, NA will be treated as Zero.
# If the data are all NA, then the result will be NA.
print (df['one'].sum())

# Replace NAN with scalar value
print ("NaN replaced with '0':")
print (df.fillna(0))
print("--------")

# Fill NA Forward and Backward
print("-------fill forward------")
print (df.fillna(method='pad'))                                       # Fill methods Forward  ('fill' can also be used)
print("-------fill backward------")
print (df.fillna(method='backfill'))                                  # Fill methods backward ('bfill' can also be used)
print("----------")

# Drop NA values
# By default, axis=0, i.e., along row, which means that if any value within a row is NA then the whole row is excluded.
print (df.dropna())
print("-----------")
print (df.dropna(axis=1))                                              
```

```python
# GROUPBY 
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

print (df.groupby('Team'))                                 #groupby
print (df.groupby('Team').groups)                          #view groups
print("------------")
print (df.groupby(['Team','Year']).groups)                 #group by multiple columns
print("------------")
grouped = df.groupby('Year')
for name,group in grouped:                                 #iterating through groups
    print (name)
    print (group)
print("--------------")
print (grouped.get_group(2014))                            # select a group                                                                

print (grouped['Points'].agg(np.mean))                     # Using aggregation
print("---------------")
print ("size = ",grouped.agg(np.size))                     # Size of each group
print("-----------------")

grouped = df.groupby('Team')                               # Applying multiple aggregate functions
print (grouped['Points'].agg([np.sum, np.mean, np.std]))

score = lambda x: (x - x.mean()) / x.std()*10              # Using transformations
print (grouped.transform(score))

print (df.groupby('Team').filter(lambda x: len(x) >= 3))   # Using filters
```

```python
# MERGING : used for joining dataframes  
# Parameters used in MERGE function :
# 1. left − A DataFrame object.
# 2. right − Another DataFrame object.
# 3. on − Columns (names) to join on. Must be found in both the left and right DataFrame objects.
# 4. left_on − Columns from the left DataFrame to use as keys. Can either be column names or arrays with length equal to the length of the DataFrame.
# 5. right_on − Columns from the right DataFrame to use as keys. Can either be column names or arrays with length equal to the length of the DataFrame.
# 6. left_index − If True, use the index (row labels) from the left DataFrame as its join key(s). In case of a DataFrame with a MultiIndex (hierarchical), the number of levels must match the number of join keys from the right DataFrame.
# 7. right_index − Same usage as left_index for the right DataFrame.
# 8. how − One of 'left', 'right', 'outer', 'inner'. Defaults to inner. Each method has been described below.
# 9. sort − Sort the result DataFrame by the join keys in lexicographical order. Defaults to True, setting to False will improve the performance substantially in many cases.


left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print (left)
print (right)

print (pd.merge(left,right,on='id'))                               # Merge two dataframes on id
print (pd.merge(left,right,on=['id','subject_id']))                # Merge Two DataFrames on Multiple Keys

# 'how' argument
# left - left outer join,   right - right outer join,   outer - full outer join,inner - inner join 

print("----------- LEFT JOIN -------------") 
print (pd.merge(left, right, on='subject_id', how='left'))            # Left join
print("------------ RIGHT JOIN --------------")
print (pd.merge(left, right, on='subject_id', how='right'))           # right join
print("------------ OUTER JOIN ---------------")
print (pd.merge(left, right, how='outer', on='subject_id'))            # outer join
print("------------ INNER JOIN ---------------")
print (pd.merge(left, right, on='subject_id', how='inner'))            # inner join
```

```python
# CONCATENATION
# Consist of following parameters: 
# 1. objs − This is a sequence or mapping of Series, DataFrame, or Panel objects.
# 2. axis − {0, 1, ...}, default 0. This is the axis to concatenate along.
# 3. join − {‘inner’, ‘outer’}, default ‘outer’. How to handle indexes on other axis(es). Outer for union and inner for intersection.
# 4. ignore_index − boolean, default False. If True, do not use the index values on the concatenation axis. The resulting axis will be labeled 0, ..., n - 1.
# 5. join_axes − This is the list of Index objects. Specific indexes to use for the other (n-1) axes instead of performing inner/outer set logic.

one = pd.DataFrame({
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5'],
   'Marks_scored':[98,90,87,69,78]},
   index=[1,2,3,4,5])

two = pd.DataFrame({
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5'],
   'Marks_scored':[89,80,79,97,88]},
   index=[1,2,3,4,5])
print (pd.concat([one,two]))                                                  # Works like UNION
print("------------------")
print (pd.concat([one,two],keys=['x','y']))                                   # specific keys associated with each of the pieces of the chopped up DataFrame                       
print("--------------------")
print (pd.concat([one,two],keys=['x','y'],ignore_index=True))                 # combine index for all data.
print("-------------------")
print (pd.concat([one,two],axis=1))                                           # If two objects need to be added along axis=1, then the new columns will be appended.
print("-------------------")
print (one.append(two))                                       # Concate using append function. Concate along axis=0
print (one.append([two,one,two]))                                             # Multiple objects append
```

```python
# CATEGORICAL DATA
# Categorical variables can take on only a limited, and usually fixed number of possible values. Besides the fixed length, categorical data might have an order but cannot perform numerical operation. Categorical are a Pandas data type.
# Data includes the text columns, which are repetitive. Features like gender, country, and codes are always repetitive. These are the examples for categorical data.

# object creation
s = pd.Series(["a","b","c","a"], dtype="category")                       # By specifying the dtype as "category" in pandas object creation.
print (s)
print("----------")

cat = pd.Categorical(['a', 'b', 'c', 'a', 'b', 'c'])                     # pandas.Categorical(values, categories, ordered)  
print (cat)                                                              # Using the standard pandas Categorical constructor
print("-----------")
cat = cat=pd.Categorical(['a','b','c','a','b','c','d'], ['c', 'b', 'a'])     # any value which is not present in the categories will be treated as NaN.
print (cat)
print("------------")
cat = cat=pd.Categorical(['a','b','c','a','b','c','d'], ['c', 'b', 'a'],ordered=True)         
print (cat)
print("--------describe---------")
cat = pd.Categorical(["a", "c", "c", np.nan], categories=["b", "a", "c"])
df = pd.DataFrame({"cat":cat, "s":["a", "c", "c", np.nan]})
print (df["cat"].describe())                                                # describe

print (cat.categories)                                                      # used to get the categories of the object.
print("--------")
print (cat.ordered)                                                         # used to get the order of the object. return false if we haven't specified any order.
print("--------")
#Renaming categories
s = pd.Series(["a","b","c","a"], dtype="category")
s.cat.categories = ["Group %s" % g for g in s.cat.categories]
print (s.cat.categories)
print("----------")
# Appending category
s = s.cat.add_categories([4])
print (s.cat.categories)
print("----------")
# Removing category
s = pd.Series(["a","b","c","a"], dtype="category")
print ("Original object:")
print (s)
print ("After removal:")
print (s.cat.remove_categories("a"))
print("----------")
# Comparing categorical data
cat = pd.Categorical([1,2,3], [1,2,3],ordered=True)
cat1 = pd.Categorical([2,2,2], [1,2,3],ordered=True)

print (cat>cat1)
# Comparison is possible only in following situations:
# 1. comparing equality (== and !=) to a list-like object (list, Series, array, ...) of the same length as the categorical data.
# 2. all comparisons (==, !=, >, >=, <, and <=) of categorical data to another categorical Series, when ordered==True and the categories are the same.
# 3. all comparisons of a categorical data to a scalar.
```

```python
# SPARSE DATA
# Sparse objects are “compressed” when any data matching a specific value (NaN / missing value, though any value can be chosen)
# is omitted. A special SparseIndex object tracks where data has been “sparsified”. This will make much more sense in an example.
# All of the standard Pandas data structures apply the to_sparse method 

ts = pd.Series(np.random.randn(10))                         
ts[2:-2] = np.nan
sts = ts.to_sparse()                                        # The sparse objects exist for memory efficiency reasons.
print (sts)
print("------------")

df = pd.DataFrame(np.random.randn(10000, 4))
df.ix[:9998] = np.nan
sdf = df.to_sparse()
print (sdf.density)
print("-------------")

ts = pd.Series(np.random.randn(10))                         # Any sparse object can be converted back to the standard dense form by calling to_dense 
ts[2:-2] = np.nan
sts = ts.to_sparse()
print (sts.to_dense())
print("-------------")


# SPARSE DATA TYPE
# Sparse data should have the same dtype as its dense representation. Currently, float64, int64 and booldtypes are supported.
# Default values : float64 − np.nan, int64 − 0, bool − False

s = pd.Series([1, np.nan, np.nan])
print (s)
s.to_sparse()
print (s)
```
